```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\moseq-extract-testing\test_2animal_full_timecourse\20MOS08-F\session_20201026124130\proc_5k\keypoints_00.tsv", sep='\t')
df
```

```python
kpts = [
    'Nose',
    'Neck',
    'Left Ear',
    'Right Ear',
    'Left Hip',
    'Right Hip',
    'TailBase',
    'TailTip'
]
fig, axs = plt.subplots(len(kpts) // 2, 2, sharex=True, sharey=True, figsize=(20,20))
for kp, ax in zip(kpts, axs.flat):
    sns.histplot(x=df[f'reference/{kp}_score'], ax=ax)
    ax.set_title(f'Keypoint "{kp}" Score')
    ax.set_xlabel('Model Confidence Score')
```
